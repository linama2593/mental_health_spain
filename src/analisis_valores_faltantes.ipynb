Funciones con utilidades

```python
import json

def cambiar_nombres_desde_json(dataframe, ruta_json):
    # Compruebo que puedo cargar el JSON
    with open(ruta_json, 'r') as file:
        nombres_nuevos = json.load(file)

    df_con_nombres = dataframe.copy()  # Crear una copia del DataFrame para evitar modificar el original
    
    for columna in dataframe.columns:
        nuevo_nombre = nombres_nuevos.get(columna, {}).get('Descripción', columna)
        df_con_nombres.rename(columns={columna:f'{columna};{nuevo_nombre}' }, inplace=True)

    return df_con_nombres
```

### Analisis de valores faltantes

```python
columnas_con_nan = df.columns[df.isna().any()].tolist()

df_faltantes = df[columnas_con_nan]

df_faltantes

```

```python
print(round(len(df_faltantes.columns) / len(df.columns) * 100 ,2) , '% de columnas con valores faltantes') 
```

```python
df_faltantes_completos = cambiar_nombres_desde_json(df_faltantes, '../data/json_files/var_dict.json')
```

```python
pd.set_option('display.max_rows', None)
```

```python
# Contar la cantidad de valores faltantes por cada columna
valores_faltantes_por_columna = df_faltantes_completos.isna().sum()

# Imprimir la cantidad de valores faltantes por cada columna
valores_faltantes_por_columna
```

```python
valores_faltantes_ordenados = valores_faltantes_por_columna.sort_values(ascending=False)

# Redirigir la salida estándar al archivo
with open('../assets/columnas_con_missings.txt', 'w') as file:
    print(valores_faltantes_ordenados, file=file)
```
